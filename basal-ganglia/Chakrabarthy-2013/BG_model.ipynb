### Packages import

```python
from dana import *
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Helper functions

```python
def H(x):
    return 1 if x > 0 else 0
   
```

### Simulation parameters

```python
# Population size
n = 20

# Default Time resolution
dt = 1.0*millisecond

# Default trial duration
duration = 500

# Initialization of the random generator (reproductibility !)
np.random.seed(1)
```

```python
# Time constants
tau_GPe = 10 * dt
tau_STN = 10 * dt
tau_GPi = 10 * dt
tau     = 10 * dt
```

```python
theta_D1  =0.1
theta_D2  = -0.1
theta_GPe = 3
theta_STN = -3
```

```python
# Dopamine levels
delta = -2.
```

```python
# Compute of parameters
lamda_D1  = 5.*(1/(1+np.exp(-6.*(delta - theta_D1))))
lamda_D2  = 5.*(1/(1+np.exp(-6.*(theta_D2 - delta))))
lamda_GPe = 4. * H(theta_GPe - delta) + 1  
lamda_STN = 4. * H(theta_STN - delta) + 1
```

```python
k_x   = (2.*np.pi)/n                                
A     = 10.
sigma =1.2
C     = 0.1

d = np.zeros((n,n))
for i in range(0,n):
    a = np.array([i,i])
    for j in range(0,n):
        b = np.array([i,j])
        d[i, j] = np.linalg.norm(a-b)
        
W_Gpi = A * np.exp(-d**2/(2*sigma**2)) - C

#parameters A, sigma, Ce{0.1,0.3}, k_x, tau are from the article from Standage 
```

```python
s = 1.0 
w = 1.0
```

### Populations

```python
W_GPe     = (-1.) * s * np.ones((n*n, n*n))
W_GPe_Str = np.ones((1,n))
W_STN     = s * np.ones((n*n, n*n))
W_GPi_STN = np.ones((1,n)) * 1./n
```

```python
I_Cx_Str = np.zeros((n,1))
I_Cx_Str[3] = 0.05
I_Cx_Str[13] = 0.2

Str1_u = np.zeros((n,1))
Str1_V = np.zeros((n,1))

Str2_u = np.zeros((n,1))
Str2_V = np.zeros((n,1))

GPe_x = np.zeros((n,n))
GPe_U = np.zeros((n,n))

STN_x = np.zeros((n,n))
STN_U = np.zeros((n,n))

GPi_I = np.zeros((n,1))
GPi_S = np.zeros((n,1))
GPi_u = np.zeros((n,1))
```

```python
for i in range(duration):
    
    Str1_u = Str1_u + (-Str1_u + Str1_V + I_Cx_Str) * dt
    Str1_V = np.tanh(lamda_D1 * Str1_u)
    
    Str2_u = Str2_u + (-Str2_u + Str2_V + I_Cx_Str) * dt
    Str2_V = np.tanh(lamda_D2 * Str2_u)
    
    GPe_x  = GPe_x + ( -GPe_x + np.dot(W_GPe, GPe_U.reshape((n*n,1))).reshape((n,n)) + w * STN_U + np.dot(Str2_V,W_GPe_Str))/tau_GPe * dt
    GPe_U  = np.tanh(lamda_GPe * GPe_x)
    
    
    STN_x  = STN_x + ( -STN_x + np.dot(W_STN, STN_U.reshape((n*n,1))).reshape((n,n)) - w * GPe_U)/tau_STN * dt
    STN_U  = np.tanh(lamda_STN * STN_x)
    
    
    GPi_I  = Str1_V + np.dot(W_GPi_STN, STN_U)
    GPi_u  = GPi_u + ((-GPi_u + k_x * np.dot(W_Gpi, GPi_S) + GPi_I) / tau_GPi) * dt
    GPi_S  = GPi_u**2 / (1 + ((1./2) * np.sum(GPi_u**2 * k_x))) 
    
```

```python
#ax = fig.add_subplot(111, projection='3d')
x = np.linspace(0,19,20)
y = np.linspace(0,19,20)
X,Y = np.meshgrid(x,y)
plt.ion()
fig = plt.figure(figsize = (8,12))
ax1 = plt.subplot2grid((5,2), (0,0), colspan=2)
ax2 = plt.subplot2grid((5,2), (1,0))
ax3 = plt.subplot2grid((5,2), (1,1))
ax4 = plt.subplot2grid((5,2), (2,0), colspan=2, projection='3d')
ax5 = plt.subplot2grid((5,2), (3,0), colspan=2, projection='3d')
ax6 = plt.subplot2grid((5,2), (4,0), colspan=2)

ax1.set_title('Input')
ax1.plot(I_Cx_Str)
ax2.set_title('D1')
ax2.plot(Str1_V)
ax2.set_ylim(-1., 1.)

ax3.set_title('D2')
ax3.plot(Str2_V)
ax3.set_ylim(-1., 1.)

ax4.set_title('GPe')
ax4.plot_surface(X, Y, GPe_U)

ax5.set_title('STN')
ax5.plot_surface(Y,X, STN_U)
#ax5.set_zlim(-1,-1)

ax6.set_title('GPi')
ax6.plot(GPi_S)
#ax6.set_ylim(0., 20.)
plt.draw()
fig.tight_layout()
plt.show()
plt.ioff()
```
